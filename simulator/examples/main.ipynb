issue token

```python
import requests
from requests.auth import HTTPBasicAuth

# Configuration
AUTH_SERVICE_URL = "http://localhost/api"  # Update with your backend URL
USERNAME = "admin" # Update with credentials in .env in trec-usersim-infra
PASSWORD = "pass"
TEAM_NAME = "team"  

# Issue token
response = requests.get(
    f"{AUTH_SERVICE_URL}/auth/issue-token",
    params={"name": TEAM_NAME},
    auth=HTTPBasicAuth(USERNAME, PASSWORD),
)

if response.status_code == 200:
    token = response.json()["token"]
    print(f"Token issued successfully: {token}")
else:
    print(f"Error: {response.status_code} - {response.json()}")
```

start run

```python
import requests

# config
BASE_URL = "http://localhost/api"  
RUN_ROUTER_BASE = f"{BASE_URL}"
AUTH_TOKEN = token  


def get_headers() -> dict:
    """Return headers with authentication token."""
    return {
        "Authorization": f"Bearer {AUTH_TOKEN}",
        "Content-Type": "application/json"
    }


def create_run(run_id: str, description: str, team: str) -> dict:
    """Create a new run instance."""
    response = requests.post(
        f"{RUN_ROUTER_BASE}/run/start",
        headers=get_headers(),
        json={
            "run_id": run_id,
            "description": description,
            "team_id": team
        }
    )
    response.raise_for_status()
    return response.json()


def main():
    run = create_run(run_id="run", description="random", team="team")
    # run_id = run["id"]
    print(f"Created run: {run}")


if __name__ == "__main__":
    main()
```

```python
import requests

# config
BASE_URL = "http://localhost/api"  
RUN_ROUTER_BASE = f"{BASE_URL}"
AUTH_TOKEN = token 


def get_headers() -> dict:
    """Return headers with authentication token."""
    return {
        "Authorization": f"Bearer {AUTH_TOKEN}",
        "Content-Type": "application/json"
    }


def continue_run(run_id: str, user_utterance: str) -> dict:
    """Continue completing the run. """
    response = requests.post(
        f"{RUN_ROUTER_BASE}/run/continue",
        headers=get_headers(),
        json={    
            "run_id": run_id,
            "user_utterance": user_utterance,
        }
    )
    response.raise_for_status()
    return response.json()


def main():
    run = continue_run(run_id="run", user_utterance="hi.")
    print(f"Continued run: {run}")


if __name__ == "__main__":
    main()
```
